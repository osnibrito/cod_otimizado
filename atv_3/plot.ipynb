```python
import pandas as pd

df = pd.read_csv("stats.csv")
df.head()
```

```python
spectral_stat = df[df["program"] == "spectral"]
n_body_stat = df[df["program"] == "n_body"]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_compile_flag(df, program):
    plt.figure(figsize=(7,4))
    ax = sns.barplot(df, x="flag", y="media_compile")
    ax.set(xlabel= "FLAG", ylabel="MEDIA DE COMPILAÇÃO (ms)", title=f"Media do tempo de compilação para cada flag - {program}")

    plt.show()

plot_compile_flag(n_body_stat, "n_body")
plot_compile_flag(spectral_stat, "spectral form")
```

```python
def plot_run_flag(df, program):
    plt.figure(figsize=(7,4))
    ax = sns.barplot(df, x="flag", y="media_run")
    ax.set(xlabel= "FLAG", ylabel="MEDIA DE EXECUÇÃO (ms)", title=f"Media do tempo de execução para cada flag - {program}")

plot_run_flag(n_body_stat, "n_body")
plot_run_flag(spectral_stat, "spectral form")
```

```python
df_full = pd.read_csv("datatime.csv")
spectral_full = df_full[df_full["program"] == "spectral"]
n_body_full = df_full[df_full["program"] == "n_body"]
```

```python
def plot_compile(df, program):
    fig, axes = plt.subplots(5, 1,figsize=(10,15))
    fig.suptitle(f"Distribuição do tempo de compilação - {program}\n")

    for counter, flag in enumerate(df["flag"].unique()):
        sns.histplot(ax=axes[counter], data=df[df["flag"]==flag], x="time_compile", bins=20)
        axes[counter].set(ylabel="Quantidade" ,xlabel="Tempo de compilação (ms)", title=f"{flag}")

    fig.tight_layout()
    plt.show()

plot_compile(spectral_full, "spectral form")
plot_compile(n_body_full, "n_body")
```

```python
def plot_run(df, program):
    fig, axes = plt.subplots(5, 1,figsize=(10,15))
    fig.suptitle(f"Distribuição do tempo de execução - {program}\n")

    for counter, flag in enumerate(df["flag"].unique()):
        sns.histplot(ax=axes[counter], data=df[df["flag"]==flag], x="time_run", bins=20)
        axes[counter].set(ylabel="Quantidade" ,xlabel="Tempo de execução (ms)", title=f"{flag}")

    fig.tight_layout()
    plt.show()

plot_run(spectral_full, "spectral form")
plot_run(n_body_full, "n_body")
```
